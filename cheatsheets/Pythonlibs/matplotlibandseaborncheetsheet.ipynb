```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Matplotlib Cheat Sheet

## 1. Importing Matplotlib
import matplotlib.pyplot as plt

## 2. Basic Plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()

## 3. Figure and Axes
fig, ax = plt.subplots()
ax.plot(x, y)

## 4. Plot Types
plt.plot(x, y, label='Line')           # Line plot
plt.scatter(x, y, label='Scatter')     # Scatter plot
plt.bar(x, y, label='Bar')             # Bar plot
plt.hist(y, bins=5, label='Histogram') # Histogram
plt.boxplot(y)                         # Boxplot
plt.pie(y, labels=x)                   # Pie chart

## 5. Customizing Plots
plt.title('Title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 35)
plt.xticks([1,2,3,4], ['A','B','C','D'])
plt.yticks([10,20,25,30])

## 6. Line Styles and Markers
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=8, markerfacecolor='blue')

## 7. Subplots
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 1].bar(x, y)
axs[1, 0].scatter(x, y)
axs[1, 1].hist(y)

## 8. Saving Figures
plt.savefig('figure.png', dpi=300, bbox_inches='tight')

## 9. Annotations
plt.annotate('Max', xy=(4, 30), xytext=(3, 28), arrowprops=dict(facecolor='black', shrink=0.05))

## 10. Colors
plt.plot(x, y, color='green')
plt.plot(x, y, c='#FFDD44')
plt.plot(x, y, c=(0.1, 0.2, 0.5))

## 11. Multiple Lines
plt.plot(x, y, label='First')
plt.plot(x, [i*2 for i in y], label='Second')
plt.legend()

## 12. Log Scale
plt.xscale('log')
plt.yscale('log')

## 13. Twin Axes
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y, 'g-')
ax2.plot(x, [i*2 for i in y], 'b-')

## 14. Show/Close/Clear
plt.show()      # Display plot
plt.clf()       # Clear current figure
plt.close()     # Close a figure window

## 15. Style Sheets
plt.style.use('ggplot')
plt.style.available

## 16. Advanced: Images
img = np.random.rand(10,10)
plt.imshow(img, cmap='viridis')
plt.colorbar()

## 17. Advanced: 3D Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot([1,2,3], [4,5,6], [7,8,9])

# For more, see: https://matplotlib.org/stable/gallery/index.html
```

```python
import seaborn as sns

# Seaborn Cheat Sheet


# 1. Load Example Dataset
tips = sns.load_dataset('tips')

# 2. Basic Plot Types

## Scatter Plot
sns.scatterplot(data=tips, x='total_bill', y='tip')

## Line Plot
sns.lineplot(data=tips, x='total_bill', y='tip')

## Bar Plot (categorical)
sns.barplot(data=tips, x='day', y='total_bill')

## Count Plot (counts of unique values)
sns.countplot(data=tips, x='day')

## Box Plot
sns.boxplot(data=tips, x='day', y='total_bill')

## Violin Plot
sns.violinplot(data=tips, x='day', y='total_bill')

## Strip Plot (jittered scatter for categorical)
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True)

## Swarm Plot (non-overlapping points)
sns.swarmplot(data=tips, x='day', y='total_bill')

## Histogram
sns.histplot(data=tips, x='total_bill', bins=20)

## KDE Plot (Kernel Density Estimate)
sns.kdeplot(data=tips['total_bill'], fill=True)

# 3. Relational Plots

## relplot (scatter or line, with facets)
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', col='time', kind='scatter')

# 4. Categorical Plots

## catplot (bar, box, violin, etc., with facets)
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex', col='time')

# 5. Matrix Plots

## Heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')

## Clustermap
sns.clustermap(corr, annot=True, cmap='viridis')

# 6. Pairwise Plots

## Pairplot (scatterplot matrix)
sns.pairplot(tips, hue='sex')

# 7. Regression Plots

## regplot (scatter + regression line)
sns.regplot(data=tips, x='total_bill', y='tip')

## lmplot (regression with facets)
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', col='time')

# 8. Distribution Plots

## distplot (deprecated, use histplot/kdeplot)
# sns.distplot(tips['total_bill'])

# 9. FacetGrid (custom multi-plot grid)
g = sns.FacetGrid(tips, col='sex', row='time')
g.map(sns.scatterplot, 'total_bill', 'tip')

# 10. Customization

## Set Style
sns.set_style('whitegrid')  # 'darkgrid', 'white', 'dark', 'ticks'

## Set Color Palette
sns.set_palette('pastel')   # 'deep', 'muted', 'bright', 'dark', 'colorblind'

## Set Context (scaling)
sns.set_context('talk')     # 'paper', 'notebook', 'talk', 'poster'

## Add Title and Labels
ax = sns.scatterplot(data=tips, x='total_bill', y='tip')
ax.set_title('Total Bill vs Tip')
ax.set_xlabel('Total Bill ($)')
ax.set_ylabel('Tip ($)')

# 11. Save Figure
import matplotlib.pyplot as plt
plt.savefig('seaborn_figure.png', dpi=300, bbox_inches='tight')

# 12. Show Plot
plt.show()

# 13. Useful Functions

## Show available color palettes
print(sns.color_palette())

## Show available datasets
print(sns.get_dataset_names())

# 14. Advanced: Customizing Legends
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')
plt.legend(title='Legend Title', loc='upper left')

# 15. Advanced: Multiple Plots in One Figure
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=tips, x='day', y='total_bill', ax=axs[0])
sns.violinplot(data=tips, x='day', y='total_bill', ax=axs[1])

# 16. Advanced: Annotating Plots
ax = sns.barplot(data=tips, x='day', y='total_bill')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                textcoords='offset points')

# For more, see: https://seaborn.pydata.org/tutorial.html
```
